
# SparkSession & Basic DataFrame Operations

## Business Scenario
A leading digital and network services provider processes large-scale Call Detail Records (CDR) daily.
The production dataset `telecom_cdr_production.csv` contains enriched CDR attributes including:

- Revenue calculations
- Roaming multipliers
- SLA latency measurements
- SLA breach indicators
- Plan segmentation (PREPAID / POSTPAID)

## Problem Statement
Network Operations wants an initial structural analysis of incoming production CDR data to:

1. Validate schema correctness
2. Understand distribution of services
3. Inspect SLA breach patterns
4. Validate revenue calculation integrity

## Business Requirements
- Initialize SparkSession
- Read production CSV with proper schema inference
- Inspect schema and record count
- Perform basic select, filter, and column-level exploration


```python

from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.appName("Telecom_Production_CDR_Analysis").getOrCreate()

df = spark.read.csv("telecom_cdr_production.csv", header=True, inferSchema=True)

df.printSchema()
# df.show(5)
# df.count()
```

```python

# Basic exploration
df.select("service_type", "plan_type", "roaming_type", "revenue").show(10)

# SLA breach quick validation
df.filter(col("sla_breach_flag") == 1).count()
```
