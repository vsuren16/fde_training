```python
from pyspark.sql import SparkSession

import os
os.environ["PYSPARK_PYTHON"] = r"C:\Users\Administrator\AppData\Local\Programs\Python\Python310\python.exe"
os.environ["PYSPARK_DRIVER_PYTHON"] = r"C:\Users\Administrator\AppData\Local\Programs\Python\Python310\python.exe"

# If you're in a notebook, it's common to have an old SparkSession running.
# This makes the cell re-runnable safely.
spark = SparkSession.builder.master("local[2]").appName("pyspark-local").getOrCreate()
sc = spark.sparkContext

sc.setLogLevel("ERROR")

data = [1, 2, 3, 4, 5]

rdd = sc.parallelize(data, 2)
print("collect rdd:", rdd.collect())

try:
    rdd_map = rdd.map(lambda x: x ** 2)
    print("collect rdd_map:", rdd_map.collect())

    rdd_filter = rdd_map.filter(lambda x: x > 3)
    print("collect rdd_filter:", rdd_filter.collect())

    sentences = sc.parallelize(["Hello world", "PySpark is great"], 2)
    rdd_map_split = sentences.map(lambda s: s.split(" "))
    print("collect rdd_map split:", rdd_map_split.collect())

    sentences = sc.parallelize(["Hello world", "PySpark is great"], 2)
    rdd_flat = sentences.flatMap(lambda s: s.split(" "))
    print("collect rdd_flat_map:", rdd_flat.collect())

    rdd_text = sc.textFile("C:\\Users\\test.txt")  # Make sure the file path is correct
    rdd_map_text = rdd_text.map(lambda s: s.split(" "))
    rdd_map_text_count = rdd_map_text.map(lambda words: len(words))
    rdd_map_text_flat_map = rdd_map_text.flatMap(lambda words: words)
    print("Text from file: ",rdd_map_text.collect())  # Print the first line/element of the RDD
    print("Word counts: ",rdd_map_text_count.collect())
    print("Flat map result: ",rdd_map_text_flat_map.collect())
    print("Total word count: ", rdd_map_text_flat_map.count())

except Exception as e:
    # This prints the actual Java-side root cause text in most notebook setups
    print("ERROR:", e)

spark.stop()
```
