```python
@function_tool
def calculator(expression: str) -> str:
    """
    Evaluates a mathematical expression.
    Example input: "234 * 56"
    """
    try:
        result = eval(expression)
        return str(result)
    except Exception as e:
        return f"Error: {str(e)}"
 
```

```python
@function_tool
def extract_name(full_name: str) -> dict:
    """
    Extract first and second name from a full name string.
    If more than two names are provided, everything after the first
    token is treated as the second name.
    """
    parts = full_name.strip().split()

    if len(parts) < 2:
        raise ValueError("Please provide at least a first and second name.")

    first_name = parts[0]
    second_name = " ".join(parts[1:])

    return {
        "first_name": first_name,
        "second_name": second_name
    }
```

```python
agent = Agent(
    name="SimpleMathAgent",
    instructions="""
You are a helpful AI agent.
 
- Answer general knowledge questions directly.
- Use the calculator tool whenever mathematical computation is required.
- For name extraction tasks, use the extract_name tool.
- Always use the tool for arithmetic.
""",
    tools=[calculator,extract_name],
    model="gpt-4o-mini"
)
```

```python
result = await Runner.run(
    agent,
    "My name is Surendar"
)
 
print(result.final_output)
```
